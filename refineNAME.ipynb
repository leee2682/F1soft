```python
import os
from tqdm import tqdm

path = r'F:/NIA73/'

folder_lst = os.listdir(path)

nia_lst = []

for folder in tqdm(folder_lst):
    folder = path + folder + '/'
    folder_lst2 = os.listdir(folder)
    for file in folder_lst2:
        if (file[2:4] >= '13') and (file[2:4] <= '19'):
            nia_lst.append(file.split('.')[0])
```

```python
import cv2
import shutil
from tqdm import tqdm
import json
import os

path = r'G:\2. 라벨링데이터셋/'
move_path = r'H:/dd/'

for cls in tqdm(os.listdir(path)):
    for state in os.listdir(os.path.join(path, cls)):
        for file in os.listdir(os.path.join(path, cls, state)):
            src = os.path.join(path, cls, state, file)
            try:
                with open(src, "r", encoding="utf8") as f:
                    data = json.load(f)

                if data['info']['width'] <= 1000 or data['info']['height'] <= 1000:
                    print(file, data['info']['width'], data['info']['height'])
            except:
                print(file)
```
