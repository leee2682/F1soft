```python
import pandas as pd
import openpyxl
from tqdm import tqdm
import shutil
import json
import os

wb = openpyxl.Workbook()

ws = wb.active

ws['A1'] = '라벨링데이터'
ws['E1'] = '원천데이터'

def file_num(folder_path):
    folder_lst = os.listdir(folder_path)

    x = 1
    i, j = 0, 0

    for folder in tqdm(folder_lst):
        if (folder == 'None'):
            folder = folder_path + folder
            file_lst = os.listdir(folder)

            for file in file_lst:
                    i += 1
                    j += 1

            if (file.split('.')[-1] == 'json'):
                ws.cell(62, 1, 'None')
                ws.cell(62, 3, i)
            if (file.split('.')[-1] == 'jpg') or (file.split('.')[-1] == 'jpeg'):
                ws.cell(62, 5, 'None')
                ws.cell(62, 7, i)
            print('None = {}'.format(i))
            j = i - j
            i = 0
        else:
            folder = folder_path + folder
            folder_lst2 = os.listdir(folder)
    
            for folder2 in folder_lst2:
                folder2 = folder + '/' + folder2 + '/'
                file_lst = os.listdir(folder2)
            
                for file in file_lst:
                    i += 1
                    j += 1
            
                x += 1
                if (file.split('.')[-1] == 'json'):
                    ws.cell(x, 1, folder2.split('/')[-3][:2])
                    ws.cell(x, 2, folder2.split('/')[-2][:1])
                    ws.cell(x, 3, i)

                else:
                    ws.cell(x, 5, folder2.split('/')[-3][:2])
                    ws.cell(x, 6, folder2.split('/')[-2][:1])
                    ws.cell(x, 7, i)

                i = 0

    if (file.split('.')[-1] == 'json'):
        print('라벨링데이터 - {}'.format(abs(j)))
    else:
        print('원천데이터 - {}'.format(abs(j)))

file_num(r'F:\라벨링데이터/')
file_num(r'F:/원천데이터/')

wb.save(r'C:/Users/admin/Desktop/73.xlsx')
```
