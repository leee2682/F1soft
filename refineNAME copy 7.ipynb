```python
import cv2
import shutil
from tqdm import tqdm
import json
import os

path = r'F:/ss/'
move = r'F:/d/'

clas = ['01. Bollard_', '02. Bollard_', '03. _2', '04. _3', '05.  ', '06.  ', '07. ', '08. ', '09. ', '10. () ', '11. ( )', '12.  ', '13.  ', '14. ', '15.   ', '16. ', '21.  ', '23.  ', '24. ', '25. ', '26. ', '27. ', '28. ,  ', '29. () ', '30. () ', '31. Bollard_', '32. Bollard_U', '36. ', '37. _', '38. _']
clas2 = ['0. ', '1. ']

for file in os.listdir(path):
    src = os.path.join(path, file)
    dst = os.path.join(move, file)

    
    try:
        with open(src, "r", encoding="utf8") as f:
            data = json.load(f)
    except:
        print(file)
        os.rename(src, dst)
    json_split = data['info']['filename'].split('_')
    json_name2 = '{}_{}_{}_{}_{}_{}_{}_{}'.format(json_split[0], json_split[1], data['description']['state'],
                                                json_split[3], data['info']['city_id'], json_split[-3], json_split[-2],
                                                json_split[-1])
    
    json_name = json_name2.split('_')

    try:
        src = [s for s in refine if json_name2[9:] in s][0]
        img = cv2.imread(src)
        
        h, w, c = img.shape
        
        dst = os.path.join('F://{}/{}/'.format(clas[int(json_name[1]) - 1], clas2[int(data['description']['state'])]), json_name2)
        dst2 = os.path.join('F://{}/{}/'.format(clas[int(json_name[1]) - 1], clas2[int(data['description']['state'])]), '{}.json'.format(json_name2.split('.')[0]))
        
        data['info']['filename'] = json_name2
        data['info']['width'] = w
        data['info']['height'] = h
    except:
        print(file)
    try:
        with open(dst2, 'w', encoding="utf8") as outfile:
            json.dump(data, outfile, indent = '\t', ensure_ascii = False)
        os.rename(src, dst)

    except:
        print('b')
```
