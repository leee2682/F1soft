```python
import cv2
import shutil
from tqdm import tqdm
import json
import os

path = r'F:/sss/'
move = r'F:/d/'

clas = ['01. Bollard_스테인리스', '02. Bollard_탄성고무', '03. 시선유도봉_2줄', '04. 시선유도봉_3줄', '05. 보행자용 방호울타리', '06. 교량용 방호울타리', '07. 턱낮추기', '08. 경사로', '09. 점자블럭', '10. (도로안내표지) 지주', '11. 보도(시멘트 콘크리트)', '12. 보도 블록', '13. 자전거 도로', '14. 연석', '15. 무단횡단 방지 울타리', '16. 맨홀', '21. 현장 신호제어기', '23. 시각장애인용 음향신호기', '24. 과속방지턱', '25. 횡단보도', '26. 고원식횡단보도', '27. 통합표지', '28. 정주식, 부착식 표지', '29. (가로등) 지주', '30. (전봇대) 빗금표시', '31. Bollard_대리석', '32. Bollard_U자형', '36. 소화전', '37. 주차멈춤턱_블럭형', '38. 보행자_계단']
clas2 = ['0. 양호', '1. 불량']

for file in os.listdir(path):
    src = os.path.join(path, file)
    
    with open(src, "r", encoding="utf8") as f:
        data = json.load(f)
    
    json_split = data['info']['filename'].split('_')
    json_name2 = '{}_{}_{}_{}_{}_{}_{}_{}'.format(json_split[0], json_split[1], data['description']['state'],
                                                json_split[3], data['info']['city_id'], json_split[-3], json_split[-2],
                                                json_split[-1])
    
    json_name = json_name2.split('_')

    try:
        src = [s for s in refine if json_name2[9:-5] in s][0]
        
        img = cv2.imread(src)

        h, w, c = img.shape
        
        dst = os.path.join('F:/원천데이터/{}/{}/'.format(clas[int(json_name[1]) - 1], clas2[int(data['description']['state'])]), json_name2)
        dst2 = os.path.join('F:/라벨링데이터/{}/{}/'.format(clas[int(json_name[1]) - 1], clas2[int(data['description']['state'])]), '{}.json'.format(json_name2.split('.')[0]))

        data['info']['filename'] = json_name2
        data['info']['width'] = w
        data['info']['height'] = h
        
        with open(dst2, 'w', encoding="utf8") as outfile:
            json.dump(data, outfile, indent = '\t', ensure_ascii = False)
        
        try:
            os.rename(src, dst)
        except:
            print(dst)
    except:
        continue
```
